#Importing a simpletransformers library

```python
from simpletransformers.language_modeling import LanguageModelingModel,LanguageModelingArgs
```

#Define Model arguments

```python
model_args = LanguageModelingArgs()
model_args.reprocess_input_data = True
model_args.overwrite_output_dir = True
model_args.num_train_epochs = 2
model_args.best_model_dir = "outputs/best_model"
model_args.save_best_model =True
model_args.train_batch_size = 8
model_args.dataset_type = "simple"
model_args.mlm = False
model_args.vocab_size = 50257
```

#Reading a text data

```python
with open("data.txt", "r") as f:
    veriler = f.readlines()
from sklearn.model_selection import train_test_split

train_veriler, test_veriler = train_test_split(veriler, test_size=0.2)
```

#Creating train and test datas

```python
with open("train.txt", "w") as f:
    for veri in train_veriler:
        f.write(veri)

with open("test.txt", "w") as f:
    for veri in test_veriler:
        f.write(veri)
```

#Define a model with arguments

```python
train_file = "train.txt"
test_file = "test.txt"

model = LanguageModelingModel(
    "gpt2", "gpt2", args=model_args, use_cuda=False,train_files=train_file
)
```

#Training a model

```python
model.train_model(train_file, eval_file=test_file)
```

#Testing a model

```python
result= model.eval_model(test_file)
```
